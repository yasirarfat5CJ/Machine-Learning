```python
import numpy as np
```

```python
# aggregare functions
arr=np.array([1,2,3,4])
print(np.sum(arr))
print(np.mean(arr))
print(np.std(arr))
print(np.var(arr))
print(np.min(arr))
print(np.max(arr))
```

```python

mat = np.array([[1,2,3],[4,5,6]])
print(np.sum(mat, axis=0))
```

```python
#cumalative opeartions
print(np.cumsum(arr))
print(np.cumprod(arr))
```
