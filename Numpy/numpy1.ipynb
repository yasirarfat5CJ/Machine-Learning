```python
import numpy as np
import time

# Using Python list
list1 = list(range(10))
start = time.time()
list_result = [x * 2 for x in list1]
print("List time:", time.time() - start)

# Using NumPy array
arr = np.arange(10)
start = time.time()

arr_result = arr * 2
print("NumPy time:", time.time() - start)
```
