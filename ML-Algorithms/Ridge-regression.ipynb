```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
import pandas as pd
from sklearn.metrics import mean_squared_error,r2_score

```

```python
# Sample dataset
df= {
    'Area': [1000, 1500, 2000, 2500, 3000],
    'Bedrooms': [2, 3, 3, 4, 4],
    'LocationScore': [7, 8, 9, 6, 10],
    'Price': [150000, 200000, 250000, 270000, 350000]
}

data=pd.DataFrame(df)

```

```python
X=data[['Area','Bedrooms','LocationScore']]
y=data['Price']
print(X)
print(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=17)
```

```python
model=LinearRegression()
model.fit(X,y)
```

```python
pred = model.predict(X_test)  
print(pred)
print("R² Score:", r2_score(y_test, pred))
```

```python
## Ridge regression 
model1=Ridge()
model1.fit(X,y)
pred1=model1.predict(X_test)
print(pred1)
r2_ridge = r2_score(y_test, pred1)
print("Ridge R² Score:", r2_ridge)
```

```python
##Lasso regression

# Step 1: Create and fit the Lasso model
model2 = Lasso(alpha=1.0)   # you can tune alpha later
model2.fit(X_train, y_train)

# Step 2: Predict on test data
y_pred_lasso = model2.predict(X_test)

# Step 3: Evaluate model performance
r3_lasso = r2_score(y_test, y_pred_lasso)
print("Lasso R² Score:", r3_lasso)
```
