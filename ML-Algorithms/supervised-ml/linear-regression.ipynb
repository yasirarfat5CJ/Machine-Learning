```python
from sklearn.datasets import fetch_california_housing
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
df=fetch_california_housing()
```

```python
data.columns=df.feature_names
# data['Price'] = np.random.randint(50000, 500000, size=len(data))
data.head()
```

```python
data.drop('Price',axis=1,inplace=True)
data.head()
```

```python
data['Price']=df.target
data.head()
```

```python
## dividing the data set into independeb=nt and dependent features
X=data.iloc[:,:-1] ## skipping last feature
y=data.iloc[:,-1] # dependent features
```

```python
X.head()
```

```python
y.head()
```

```python
##Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=18)

linear_reg=LinearRegression()

linear_reg.fit(X_train,y_train)

from sklearn.metrics import mean_squared_error, r2_score

# Predictions
y_pred_train = linear_reg.predict(X_train)
y_pred_test = linear_reg.predict(X_test)

# Metrics
print("Training Mean Squared Error =", mean_squared_error(y_train, y_pred_train))
print("Test Mean Squared Error =",mean_squared_error(y_test, y_pred_test))

print("Training R2 Score =", r2_score(y_train, y_pred_train))
print("Test R2 Score =", r2_score(y_test, y_pred_test))


```

```python
#checking how  well the model forms using ridge regresssion
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()

params={'alpha':[1e-15,1e-10,1e-8,1e-6,1e-5,1e-2,10,20,30,40,50]}

ridge_reg=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_reg.fit(X,y)
```

```python
print(ridge_reg.best_params_)
print(ridge_reg.best_score_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso=Lasso()
params={'alpha':[1e-15,1e-10,1e-8,1e-6,1e-5,1e-2,10,20,30,40,50]}

lasso=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
lasso.fit(X,y)
```

```python
print(lasso.best_params_)
print(lasso.best_score_)
```

```python
## predict salary based on experience

#step1 import libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model  import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

#step2 : load or create a sample data

data={
    
    'Experience':[1,2,3,4,5,6],
    'Salary':[30000,40000,50000,60000,70000,80000]
}

#step3 : convert data into data frame

df=pd.DataFrame(data)

# there target var is salary so we have to preict the salary

#step4:  split the the data based in dependent and independent features
X = df[['Experience']]
y = df['Salary']

#step5: Train test split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=17)

#step6: initialize the model
model=LinearRegression()
model.fit(X_train,y_train)

#step7 : do prediction
y_pred=model.predict(X_test)


# Step 8: Evaluate
print("Predicted values:", y_pred)
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```
