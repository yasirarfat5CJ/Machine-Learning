```python
## joins means combing data frames
df1=pd.DataFrame({
    'id':[1,2,3,4],
    "name":['mohammed','yasir','arfat','fdoyhdbn'],

})
df2=pd.DataFrame({
    'id':[1,2,3,4],
    "course":['DSA','DBMS','OS','JS'],
    
})
inner_res=pd.merge(df1,df2,how='inner',on='id')
inner_res
```

```python
df2=pd.DataFrame({
    'id':[1,2,3,4,5,6],
    "course":['DSA','DBMS','OS','JS','devops','ai'],
    
})
```

```python
outer_res=pd.merge(df1,df2,how='outer',on='id')
outer_res
```

```python
left_res=pd.merge(df1,df2,how='left',on='id')
left_res
```
